##8/20/2013: ITS amplification

####Objective:
Amplify ITS for 9 samples and 1 blank.

####Materials:
* Symmetric PCR in Bioneer Pre-mix tubes  
    * Catalog #: K-2016 (which is the "NEW" formulation)   
    * Lot: ?  
* 0.75 µl (of 10 µm stock) per rxn of ITS4 primer (7.5 µl total)
* 0.75 µl (of 10 µm stock) per rxn of Leu 1 primer (7.5 µl total)
* hesperolinon samples in 1:50 dilutions

####Procedure:
1. Mix 7.5 µl of each of the two primers.
2. Add 1.5 µl of the primer mix into each Pre-mix tube.
3. Add 17 µl of sample DNA into Pre-mix tube.
4. Add 17 µl of dH20 into 1 Pre-mix tube for control.
5. Run thermal cycler: 
<table>
<tr><td>HOLD</td><td>96º C for 1 minute</td></tr>
<tr><td>AUTO x40</td><td>96º C for 10 seconds  
48º C for 30 seconds  
72º C (20 seconds +4 seconds/cycle)  
</td></tr>
<tr><td>HOLD</td><td>72º C for 7 minutes</td></tr>
<tr><td>HOLD</td><td>4º C forever</td></tr>
</table>
Cycler: MJ2  
Program: BGB ITS  
Time started: 1:30 PM  
Time removed from cycler: ~ 4:10 PM

```python
tube_labels = pd.Series([1,2,3,4,5,6,7,8,9])
samples = pd.Series(mysequences.loc['0808A3':'1003A2'].index)
dilutions = pd.Series('1:50')
pd.DataFrame({'Tube Labels':tube_labels,'Samples':samples, 'Dilutions':dilutions})
```
